```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Carica il dataframe con i valori del PE_RATIO per le azioni
df = PE_RATIO  # Aggiungi il percorso corretto al tuo file Excel



# Inizializza un dataframe vuoto per i log return dei migliori 40 titoli per ogni mese
df_selected = pd.DataFrame()

# Seleziona ogni mese i migliori 40 titoli in base ai valori del PE_RATIO
for col in df.columns[1:]:  # Itera su ogni colonna tranne la prima (considerando i mesi)
    # Calcola i log return mensili per tutti i titoli
    df['Log_Return'] = np.log(df['Close'] / df['Close'].shift(1)
    # Seleziona i migliori 40 titoli per il mese corrente
    best_stocks = df.nsmallest(40, col)['Titolo'].tolist()
    # Filtra il dataframe originale solo per i titoli selezionati
    df_monthly = df[df['Titolo'].isin(best_stocks)]
    # Aggiungi i log return mensili dei titoli selezionati al dataframe df_selected
    df_selected = pd.concat([df_selected, df_monthly])

# Considera i costi di transazione del 0,2%
df_selected['Log_Return_Costs'] = df_selected['Log_Return'] - 0.002 * np.abs(df_selected['Log_Return'])

# Crea un istogramma che rappresenta nel tempo i log return
df_selected['Time'] = pd.to_datetime(df_selected['Time'])
df_selected = df_selected.set_index('Time')
df_selected['Log_Return_Costs'].plot(kind='hist', bins=30)

# Mostra l'istogramma
plt.xlabel('Log Return')
plt.ylabel('Frequenza')
plt.title('Istogramma Log Return')
plt.show()
```
